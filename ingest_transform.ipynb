```python
import pandas as pd
import sqlite3
```

```python
# ==================================================================================
# put file into staging table - Create passengers table
# ==================================================================================
try:
    
    # establish connection to SQLite database
    conn = sqlite3.connect('titanic.db')

    # open file with pandas and pass to sqlite
    df = pd.read_csv('https://web.stanford.edu/class/archive/cs/cs109/cs109.1166/stuff/titanic.csv')
    df.to_sql('passengers', conn, if_exists='replace', index=False)
        
except sqlite3.Error as error:
    print('Error while connecting to the database:', error)
finally:
    if conn:
        conn.close()
```

```python
#============================================================================= 
# Review the first 10 records in the table
#=============================================================================

try:
    conn = sqlite3.connect('titanic.db')
    c = conn.cursor()

    # run query against our new table
    passenger_head = c.execute("SELECT * FROM passengers LIMIT 10;").fetchall()

except sqlite3.Error as error:
    print("Error while connecting to SQLite database:", error)

finally:
    if (conn):
        conn.close()

# passenger_head
columns = ['Survived', 'Pclass', 'Name', 'Sex', 'Age', 'Siblings/Spouses Aboard', 'Parents/Children Aboard', 'Fare']
pd.DataFrame(passenger_head, columns=columns)
```

```python
#============================================================================= 
# Summarise the passengers table
#=============================================================================
try:
    sql = """ 
    SELECT  COUNT(*) AS number_of_passengers,
            Survived,
            Pclass,
            Sex
    FROM    passengers
    GROUP BY Survived, Pclass, Sex;
    """
    summary_columns = ['number_of_passengers', 'Survived', 'Pclass', 'Sex']
    conn = sqlite3.connect('titanic.db')
    c = conn.cursor()
    passenger_summary = c.execute(sql).fetchall()
    
except sqlite3.Error as error:
    print("Error while connecting to SQLite database:", error)

finally:
    if (conn):
        conn.close()
    
# passenger_summary
pd.DataFrame(passenger_summary, columns=summary_columns)
```
